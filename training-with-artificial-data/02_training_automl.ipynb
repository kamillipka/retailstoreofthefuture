```python
import os

import h2o
from h2o.automl import H2OAutoML
```

```python
h2o.init()
```

```python
data_dir = 'data_0419_0'
train_file_path = os.path.join(data_dir, 'train.csv')
```

```python
train_df = h2o.import_file(train_file_path)
train_df.describe()
```

```python
y = 'coupon_used'
x = train_df.columns
x.remove(y)
train_df['coupon_used'] = train_df['coupon_used'].asfactor()
```

```python
aml = H2OAutoML(max_models=10, seed=1, balance_classes=True)
aml.train(x=x, y=y, training_frame=train_df)
```
