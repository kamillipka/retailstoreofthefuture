# AmEx Predicting Coupon Redemption - Train Using AutoML

```python
import h2o
from h2o.automl import H2OAutoML
h2o.init()
```

```python
train = h2o.import_file('final_data/final_training_data.csv')
```

```python
train.describe()
```

```python
# Factorize the target column
train['redemption_status'] = train['redemption_status'].asfactor()
```

```python
y = 'redemption_status'
x = train.columns
x.remove(y)
```

## Run AutoML

```python
aml = H2OAutoML(max_models=10, seed=1)
aml.train(x=x, y=y, training_frame=train)
```

```python
lb = aml.leaderboard
lb.head()
```

```python
h2o.save_model(aml.leader, path='best_model_bin')
```

```python
aml = H2OAutoML(max_models=10, seed=1, balance_classes=True)
aml.train(x=x, y=y, training_frame=train)
```

```python
lb = aml.leaderboard
lb.head()
```

```python
aml = H2OAutoML(max_models=10, seed=1, balance_classes=True, sort_metric='AUCPR')
aml.train(x=x, y=y, training_frame=train)
```

```python
aml.leaderboard.head()
```

```python
aml.leader
```

### Predictions on the test set

```python
test = h2o.import_file('final_data/final_test_data.csv')
```

```python
preds = aml.leader.predict(test)
```

```python
preds.as_data_frame().predict.value_counts(normalize=True)
```
